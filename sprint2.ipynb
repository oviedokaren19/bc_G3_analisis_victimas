# Task
Analyze the `df_filtrado` dataframe to identify the most frequent `HECHO` for each of the top `ESTADO_DEPTO` values identified in a previous step. Suggest a suitable visualization for the results.

## Filter data

### Subtask:
Filter the `df_filtrado` DataFrame to include only the rows where `ESTADO_DEPTO` is in the list of `top_estados`.


**Reasoning**:
Filter the dataframe to include only the top states and display the head of the filtered dataframe.



```python
top_estados_list = top_estados['ESTADO_DEPTO'].tolist()
df_top_states = df_filtrado[df_filtrado['ESTADO_DEPTO'].isin(top_estados_list)]
display(df_top_states.head())
```

## Group and count

### Subtask:
Group the filtered DataFrame by `ESTADO_DEPTO` and `HECHO` and count the occurrences of each combination.


**Reasoning**:
Group the filtered DataFrame by ESTADO_DEPTO and HECHO, count the occurrences, convert to DataFrame, reset index, and rename the count column to 'Count'.



```python
hecho_counts = df_top_states.groupby(['ESTADO_DEPTO', 'HECHO']).size().to_frame(name='Count').reset_index()
display(hecho_counts.head())
```

## Identify top hechos per state

### Subtask:
For each of the top states, find the `HECHO` with the highest count.


**Reasoning**:
For each state in the `hecho_counts` dataframe, find the row with the maximum count, which represents the most frequent hecho for that state.



```python
top_hechos_per_state = hecho_counts.loc[hecho_counts.groupby('ESTADO_DEPTO')['Count'].idxmax()]
display(top_hechos_per_state)
```

## Visualize results

### Subtask:
Create a bar chart to visualize the top `HECHO` for each of the top states.


**Reasoning**:
Create a bar chart to visualize the top HECHO for each of the top states using the top_hechos_per_state dataframe.



```python
plt.figure(figsize=(15, 7))
sns.barplot(x='ESTADO_DEPTO', y='Count', hue='HECHO', data=top_hechos_per_state)
plt.title('Top Hecho per State')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Summary:

### Data Analysis Key Findings

* "Desplazamiento forzado" (forced displacement) is the most frequent `HECHO` across all the listed top states, with counts ranging from 4 in Casanare to 62 in Antioquia.
* The bar chart visually confirms that "Desplazamiento forzado" represents the top `HECHO` for each of the selected states.

### Insights or Next Steps

* Further analysis could investigate the factors contributing to the high frequency of "Desplazamiento forzado" in these states.
* A comparison of the counts of "Desplazamiento forzado" with other types of `HECHO` within these states could provide a clearer picture of the severity of this issue.
